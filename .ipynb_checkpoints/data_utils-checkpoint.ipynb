# Generate data file

Generate a dictionary pickle file:

    { question # : 'URL' }

```python
from bs4 import BeautifulSoup
```

```python
with open('problemList.html', 'r') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
problemList = soup.find(id="problemList")
```

```python
rows = problemList.find_all("tr")[1:]
print "Number of questions:", len(rows)
print "Example content: ", rows[0].contents
```

```python
pairs = {}
for row in rows:
    num = int(row.contents[3].text)
    url = row.contents[5].a["href"]
    isPremium = True if row.contents[5].i else False
    pairs[num] = (url, isPremium)
```

```python
pairs
```

```python
import pickle
pickle.dump(pairs, open("problemUrl.pk", "wb"))
```
